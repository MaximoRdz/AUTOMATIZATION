# Analysis of the weather scrapping

```python
with open('text_weather.txt', 'r') as f: 
    text = f.readlines()
```

```python
text = [elem.replace('\n', '').replace('', '') for elem in text] # reformat the elements
```

```python
# extract the indexes where we have a date
# TODO: generalized to every possible month and date of the year
# probably using time or datetime
day_index = [i for i, day in enumerate(text) if ('FEB' in day) or ('MAR' in day)]
day_index 
```

```python
keys = ['day', 'high', 'low']
values = np.array([text[ind:ind+3] for ind in day_index]) # numpy so we can access to it in a more confortable matrix-wise vay 
weather = {key:values[:, i] for i, key in enumerate(keys)}
```

```python
df = pd.DataFrame(weather) # show and save in a data frame
df
```
